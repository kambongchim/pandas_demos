```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_ProfitMonth = df[["month_number", "total_profit"]]
df_ProfitMonth.plot(kind = "bar", x = "month_number", y = "total_profit", title ="Company profit per month", rot = 0, ylim = (100000,500000), legend = False)
plt.grid(True, color = "grey")
plt.xticks([1,3,5,7,9,11])
plt.xlabel("Month Number")
plt.ylabel("Total profit")
```

```python
df_ProfitMonth.plot(kind = 'line', x = "month_number", y = "total_profit", 
title= "Company profit per month", marker = "o", mfc = "black", color = "red", lw = 3, linestyle = "--", ylim = (100000,500000), figsize=(8,6), label = "Profit Data of Last Year")
# plt.xticks(list(df_ProfitMonth))
plt.legend(loc = "lower right")
plt.xticks([i for i in range(1,13)])
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

```python
df_products = df[["month_number", "facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]]
df_products.plot(kind="line", x = "month_number", figsize=(6,6), marker = "o")
plt.xticks([i for i in range(1,13)])
plt.xlabel("Month Number")
plt.ylabel("Sales Unit in Number")
```

```python
df_bathingsoap = df[["month_number", "bathingsoap"]]
df_bathingsoap.plot(kind = "scatter", x = "month_number", y = "bathingsoap")
plt.grid(True, linestyle = "-")
plt.xticks([i for i in range(1,13)])
plt.xlabel("Month Number")
plt.ylabel("Number of Units Sold")
```

```python
date=["25/12","26/12","27/12"]
temp=[8.5,10.5,6.8]

df = pd.DataFrame({"Date": date, "Temp": temp})
df.plot(kind = "line", x="Date", y = "Temp", title= "Date-wise Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature")
```

```python
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]


df = pd.DataFrame({"height": height, "weight": weight})
df.plot(kind="line", x = "weight", y = "height", title ="Average weight with respect to average height.", 
marker = "o",ms = 10, mfc = "green", linestyle = "-.", color = "green")
plt.xlabel("Height in cm")
plt.ylabel("Weight in kg")
plt.legend(loc = "lower right")
```
